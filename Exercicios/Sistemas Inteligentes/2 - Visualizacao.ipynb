Intervalo de confiança

```python
from numpy import mean, std, sqrt
from scipy import stats

m1Values = [11, 13, 13, 13, 13, 13, 9, 13, 13, 13, 9, 13, 13, 13, 13, 13, 13, 13, 13, 13, 9, 13, 11, 13, 9, 11, 11, 13, 11, 11]
mean1 = mean(m1Values)
sigma1 = std(m1Values)
print ('method m1: ')
print ('method 1 mean: ', mean1) 
print ('method 1 standard deviation: ', sigma1)
print ('95% confidence interval method 1:',
       stats.norm.interval(0.95, loc=mean1, 
                           scale=sigma1/sqrt(len(m1Values))))

m2Values = [11, 13, 13, 6, 11, 11, 9, 13, 13, 11, 13, 11, 13, 13, 6, 11, 11, 11, 13, 13, 6, 13, 13, 11, 13, 6, 13, 9, 13, 13]
mean2 = mean(m2Values)
sigma2 = std(m2Values)
print ('method m2: ')
print ('method 2 mean: ', mean2) 
print ('method 2 standard deviation: ', sigma2)
print ('95% confidence interval method 2:',
       stats.norm.interval(0.95, loc=mean2, 
                           scale=sigma2/sqrt(len(m2Values))))
```

### Exercícios de Fixação

1. Obtenha média, desvio padrão e intervalo de confiança de cada característica da base iris. Contraste os resultados com os obtidos fazendo um gráfico de barras com intervalo de confiança.

```python
import numpy as np
from scipy import stats
import seaborn as sns

iris = sns.load_dataset("iris")

for s in ['sepal_length', 'sepal_width', 'petal_length','petal_width']:
    nr_Mean = np.mean(iris[s])
    nr_Std = np.std(iris[s]) 
    nr_Confidence =  stats.norm.interval(0.95, loc=np.mean(iris[s]), 
                               scale=np.std(iris[s])/np.sqrt(len(iris[s])))

    print(s)
    print('Mean: ', nr_Mean)
    print('Std: ', nr_Std)
    print('95% confidence interval method: ', nr_Confidence)
    print('\n')

# ,sepal_width,petal_length,petal_width
# output = sns.barplot(x="sepal_length", y="sepal_width", hue="species", data=iris)
```
