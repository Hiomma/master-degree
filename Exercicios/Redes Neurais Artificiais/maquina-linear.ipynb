## Maquina Linear

```python
import numpy as np
from sklearn import datasets
from sklearn.preprocessing import OneHotEncoder
```

## Exemplo 2.9

### Valores X e y

```python
X = np.array([[1,1], [2,3], [3, 1], [4,4], [2,3], [4,4], [6,1], [7,7], [6,3], [10, 10]])
y = np.array([8.00, 16.66, 14.64, 23.93, 18.99, 24.76, 19.12, 39.56, 28.36, 53.69])
```

### Monte a matriz de dados X de dimensão n×m, colocando na i-ésima linha a i-ésima entrada em forma transposta xi.T;

```python
# X transposto e seu shape
print(X.T)
print(X.T.shape)
```

### Monte a matriz de saídas desejadas Y de dimensão n×c, colocando na i-ésima linha a i-ésima saída em forma transposta yi.T;

```python
# y transposto e seu shape
print(y.T)
print(y.T.shape)
```

### Insere na matriz de dados X, antes da primeira coluna, uma coluna com o valor fixo 1 em cada linha, criando assim a matriz de dados aumentada X ̃ = [1 X] de dimensão n×(m+1);

```python
xTil = np.append(np.ones([10,1]), X, axis=1)

print(xTil.T)
print(xTil.shape)
```

### Calcule a pseudoinversa X ̃ † de dimensão (m+1)×n da matriz de dados aumentada conforme a (eq. 2.62);

```python
xInverse = np.linalg.pinv(xTil)

print(xInverse)
print(xInverse.shape)
```

### Multiplique a pseudoinversa X ̃ † de dimensão (m+1)×n pela matriz de saídas desejadas Y de dimensão n×c, resultando na matriz de pesos W de dimensão (m+1)×c, W = X ̃ †Y

```python
result = xInverse.dot(y.T)

print(result)
```

## Exemplo 2.11

```python
iris = datasets.load_iris()
X = iris.data
y = iris.target

onehot = OneHotEncoder()
yonehot = onehot.fit_transform(y.reshape(-1, 1)).todense()
```

### Monte a matriz de dados X de dimensão n×m, colocando na i-ésima linha a i-ésima entrada em forma transposta xi.T;

```python
# X transposto e seu shape
print(X.T.shape)
print(X.T)
```

### Monte a matriz de saídas desejadas Y de dimensão n×c, colocando na i-ésima linha a i-ésima saída em forma transposta yi.T;

```python
print(yonehot.shape)
print(yonehot)
```

### Insere na matriz de dados X, antes da primeira coluna, uma coluna com o valor fixo 1 em cada linha, criando assim a matriz de dados aumentada X ̃ = [1 X] de dimensão n×(m+1);

```python
xTil = np.append(np.ones([150,1]), X, axis=1)

print(xTil.T)
print(xTil.shape)
```

### Calcule a pseudoinversa X ̃ † de dimensão (m+1)×n da matriz de dados aumentada conforme a (eq. 2.62);

```python
xInverse = np.linalg.pinv(xTil)

print(xInverse)
print(xInverse.shape)
```

### Multiplique a pseudoinversa X ̃ † de dimensão (m+1)×n pela matriz de saídas desejadas Y de dimensão n×c, resultando na matriz de pesos W de dimensão (m+1)×c, W = X ̃ †Y

```python
result = xInverse.dot(yonehot)

print(result)
```
